# Mengolah Data yang Terdapat NaN

Membuat perulangan dengan variable penampung berbentuk List dengan nama 'nan_indices'. Nantinya variable ini akan menampung 'name_record' yang bernilai float

```python
nan_indices = []
for i in range(len(name_record)):
    if type(name_record[i])==float:
        nan_indices.append(i)
```

Mengubah dua variabel (yaitu "name_record" dan "time_record_total") dari tipe data yang aslinya mungkin bukan list, menjadi tipe data list.

```python
name_record = list(name_record)
time_record_total = list(time_record_total)
```

Menghilangkan nilai NaN pada variable 'name_record' dan 'time_record_total' dengan fungsi PoP, lalu munculkan nilai name_record

```python
for i in nan_indices:
    name_record.pop(i)
    time_record_total.pop(i)
name_record
```
