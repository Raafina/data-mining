# Mengimport Library

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

# Memasukan Dataset

```python
datas = pd.read_csv('healthy_lifestyle_city_2021.csv')
```

# Menyeleksi Atribut yang Dipakai

```python
data = datas[['Life expectancy(years) (Country)','Pollution(Index score) (City)','Happiness levels(Country)', 'Outdoor activities(City)']]
```

# Mengubah Objek Data kebentuk Array

```python
x = np.asarray(data)
print(x)
```

Karena terdapat nilai yang tidak sesuai, maka ganti nilai tersebut dengan nilai yang sering muncul

```python
data.loc[data['Pollution(Index score) (City)'] == '-', 'Pollution(Index score) (City)'] = '43.31'
```

# Melakukan Hierachical Klastering

```python
z = linkage(x)
print(z)
```

# Melakukan Visualisasi dengan Menggunakan Dendrogram

```python
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(z)
plt.show()
```
