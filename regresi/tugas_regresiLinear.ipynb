# Mengimport Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

# Memasukkan Dataset

```python
data = pd.read_csv('USA_Housing.csv')
x = data[['Avg. Area Income']]
y = data['Price']
```

# Split Dataset menjadi Training Set dan Testing Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

# Melakukan Fitting Simple Linear Regression pada Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

# Memprediksi hasil Test-set

```python
y_pred = regressor.predict(x_test)
```

# Visualisasi Data

```python
plt.scatter(data['Avg. Area Income'], data['Price'])
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.title('Grafik Avg. Area Income vs Price')
```

# Visualisasi Hasil Prediksi Pada Training-Set

```python
plt.figure(figsize=(10,8))
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, regressor.predict(x_train), color='red')
plt.title('Avg. Area Income to Price (Train)')
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.show()
```

# Visualisasi Hasil Prediksi Pada Test-Set

```python
plt.figure(figsize=(10,8))
plt.scatter(x_test, y_test, color = 'blue')
plt.plot(x_train, regressor.predict(x_train), color='red')
plt.title('Avg. Area Income to Price (Test Set)')
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.show()
```
